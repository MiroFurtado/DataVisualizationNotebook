```python
import sklearn
import sklearn.cluster
from sklearn.decomposition import PCA
from sklearn.cluster import AffinityPropagation
```

Read in the data and filter so that we're only left with candidates who had money.

```python
ftm_df = pd.read_pickle('df_merged')
ftm_df_havemoney =ftm_df[ftm_df['Total_$']>10**4]
numerical_ftm = ftm_df_havemoney.iloc[:,29:-2] #29-> 2 before last, these are the distributions
```

```python
scaled_mat = numerical_ftm.iloc[:,range(1,36,2)].values
for i in range(len(scaled_mat)):
    scaled_mat[i,:] /= numerical_ftm.iloc[i,-1]
mat = scaled_mat
```

```python
pca = PCA(n_components=5)
pca.fit(mat)
mat_reduced = pca.transform(mat)
#print(mat_reduced)
mat_reduced = mat_reduced[:,1:]
mat=mat_reduced
```
